EDA - Exploratory Data Analysis
EDA is essential First Step in any Data Science Project

It involves Deep look into the Dataset to understand its structure, spot pattern, identify anomalies and uncover relationships between variables.
The Goal of EDA is to get a clear picture of what the data is telling you before applying any analysis or machine learning models.
