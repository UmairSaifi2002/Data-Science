Pandas is the python library. Which make life easier for the Data Scientist.
It is Build on the numpy Wihch make Data Analysise very easy

Pandas Provide Two main Data Structure
1, Series - One dimensional array
2, Dataframe - like SQL, Excel they have Rows, Columns

```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40]) # Series is like a list with lables
print(s)
```

```python
type(s)
```

```python
s2 = pd.Series([71, 84, 56, 23, 56, 89, 98], index=["Harry", "Subh", "Rohan", "Akash", "Kirti", "John", "Rehaan"])
print(s2)
```

```python
s2["Harry"], s2["Akash"], s2["Rehaan"]
```

DataFrmae - 2D Labeled Table.
DataFrame is like a Dictionary of Series, but contains multiple columns with lables.
DataFrame Contains the data as tables in MS Excel

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Banglore"]
}
df = pd.DataFrame(data)
df
```
