### SORTIG & RANKING

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df
```

```python
# Sorting the DataFrame by an index
# df.sort_index() # by default it sorts in ascending order
# df.sort_index(ascending=False) # to sort in descending order

# Sorting the DataFrame by a single column
# df.sort_values('Year') # by default it sorts in ascending order

# Sorting the DataFrame by multiple columns
df.sort_values(['Year', 'IMDb']) # by default it sorts in ascending order
# df.sort_values('Year', ascending=False) # to sort in descending order
```

```python
df2 = df.sort_values(['Year', 'IMDb']).copy()
df2.reset_index(drop=True, inplace=True) # to reset the index after sorting, drop=True to avoid adding the old index as a column, inplace=True to modify the original DataFrame
```

### Ranking

```python
# df2['Rank'] = df2['IMDb'].rank() # by default it assigns the rank in ascending order of the columns, means the higher the value the higher the rank.
df2['Rank'] = df2['IMDb'].rank(ascending=False, method='dense') # It will give the ranking that we are using to assign the ranks
df2
```
