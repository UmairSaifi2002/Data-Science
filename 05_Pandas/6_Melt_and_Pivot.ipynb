# Melt and Pivot

```python
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Math": [85, 90, 95],
    "Science": [80, 85, 90],
    "English": [88, 92, 94]
}
df = pd.DataFrame(data)
df
```

### Using melt()

```python
df = df.melt(id_vars=['Name'], value_vars=['Math', 'Science', 'English'], var_name='Subject', value_name='Score')
df
```

```python
df = df.pivot(index='Name', columns='Subject', values='Score')
df
```
