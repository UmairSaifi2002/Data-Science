# Pie Plots

```python
import matplotlib.pyplot as plt
```

```python
names = ['Kohli', 'Dhoni', 'Rohit', 'Jadeja', 'Bumrah']
runs = [12000, 10000, 9000, 5000, 3000]
colors = ['blue', 'orange', 'green', 'red', 'purple']
explode = [0.2, 0, 0, 0, 0]  # explode the first slice (Kohli)
plt.pie(runs, labels=names, colors=colors, explode=explode, shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '--'})
plt.title('Runs scored by Indian cricketers')
```

```python
# Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.
products = ['Android', 'iOS', 'Others']
market_share = [70, 25, 5]
# Pie chart
plt.pie(market_share, labels=products, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'], wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '--'})
plt.title('Mobile OS Market Share')
```
