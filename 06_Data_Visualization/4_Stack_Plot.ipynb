# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
activities = ['Running', 'Swimming', 'Cycling', 'Yoga', 'Gym']
time_spent = [5, 3, 4, 2, 6]
colors = ['red', 'blue', 'green', 'orange', 'purple']
explode = [0, 0, 0, 0, 0.2]  # explode the first slice (Running)
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '--'})
plt.title('Time spent on different activities')
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
study_hours = [2, 3, 4, 5, 6, 7, 8]
playing_hours = [1, 2, 3, 4, 5, 6, 7]
TV_watching_hours = [1, 1, 2, 3, 4, 5, 6]
sleeping_hours = [7, 6, 5, 4, 3, 2, 1]
label = ['Study', 'Playing', 'TV Watching', 'Sleeping']
colors = ['blue', 'orange', 'green', 'red']
plt.stackplot(days, study_hours, playing_hours, TV_watching_hours, sleeping_hours, labels=label, colors=colors)
plt.title('Daily Activities Over a Week')
plt.legend(loc='upper left')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.grid()
```
