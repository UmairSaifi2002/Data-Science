```python
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3, 4, 5]
x1 = [x*2 for x in X]
x2 = [x**2 for x in X]

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(X, x1, label="Double values", marker='o', color='red') # Double the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Double Values')

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(X, x2, label="Square values", marker='s', color='blue') # Square the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Square Values')
```

```python
y1 = [x**0.5 for x in X]
y2 = [10-x for x in X]

plt.subplot(2, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(X, x1, label="Double values", marker='o', color='red') # Double the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Double Values')

plt.subplot(2, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(X, x2, label="Square values", marker='s', color='blue') # Square the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Square Values')

plt.subplot(2,2,3) # 2 rows, 2 columns, 1st plot
plt.plot(X, y1, label="Square root values", marker='^', color='green') # Square root of the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Square Root Values')

plt.subplot(2,2,4) # 2 rows, 2 columns, 2nd plot
plt.plot(X, y2, label="10 minus values", marker='d', color='orange') # 10 minus the values
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('10 minus Values')
plt.tight_layout() # Adjust layout to prevent overlap
```
