```python
import matplotlib.pyplot as plt 
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 2000, 1200, 1700, 1300, 1900, 2200]
```

```python
plt.bar(years, runs)
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs scored by a player over the years')
```

```python
# Let's compare the runs scored by Sachin, Sehwag and kohli over the years
import numpy as np
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 1900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years)) # We are generating index positions
width = 0.25 

# Plotting the bars side by side
plt.bar(x-width, sachin, width=width, label='Sachin')
plt.bar(x, sehwag, width=width, label='sehwag')
plt.bar(x+width, kohli, width=width, label='kohli')

plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Runs Scored by Sachin, Sehwag, and Kohli over the Years')
plt.legend()
plt.xticks(x, years) # Setting the x-ticks to be the years instead of index positions
plt.grid(True)
```

```python
# Horizontal Bar Plot
players = ['Sachin', 'Sehwag', 'Kohli', 'Yuvraj']
runs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
plt.barh(players, runs, color='skyblue')
plt.xlabel('Total Runs in 5 Years')
plt.ylabel('Players')
plt.title('First 5 years scores of indian batsmen.')
plt.grid(True)
```

```python
players = ['Sachin', 'Sehwag', 'Kohli', 'Yuvraj']
runs = [1500, 1200, 2400, 800]
plt.bar(players, runs, color='skyblue')
plt.xlabel('Total Runs in 5 Years')
plt.ylabel('Players')
plt.title('First 5 years scores of indian batsmen.')
for i in range(len(players)):
    plt.text(i, runs[i]+50, str(runs[i]), ha='center') 
```
