```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('Integers')
plt.ylabel('Squares')
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
Tendulkar = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]
plt.plot(years, Tendulkar)
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs scored by a player over the years')
```

```python
# Now we will plot differnt players runs in same graph
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [300, 600, 900, 1200, 1500, 1800, 2100, 2400, 1200, 0]

plt.plot(years, kohli, label='Kohli')
plt.plot(years, sehwag, label='sehwag')
plt.xlabel('years')
plt.ylabel('runs')
plt.title('Runs Scores by kohli and sehwag.')
plt.legend()
```

```python
#some new ways to plot graph
plt.plot(years, kohli, 'ro--', label='kohli')
plt.plot(years, sehwag, 'g^:', label='sehwag')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend()
plt.title('Runs scored by kohli and sehwag')
```

```python
# some easy way to tackle 
plt.plot(years, kohli, color='red', linestyle='--', label='kohli', linewidth=3)
plt.plot(years, sehwag, color='green', linestyle=':', label='sehwag')
plt.plot(years, Tendulkar, color='blue', linestyle='-.', label='Tendulkar')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend()
plt.grid(True)
plt.title('Runs scored by kohli,sehwag and tendulkar')
plt.style.use('seaborn-v0_8-deep')
```
