Slicing in Numpy Returns a View not a Copy

ðŸ”‘ Key Concepts
Copy
1, Creates a completely new array with its own memory space.
2, Changes in the copy do not affect the original array.

View
1, Shares the same underlying data buffer as the original array.
2, Only the metadata (like shape, strides, dtype) is adjusted to represent the slice.
3, Changes in the view do affect the original array, and vice versa.

Fancy Indexing

```python
arrTwo = np.array([1,9,5,2,6,7,11,40,23,67,91,39]) # this is our array
idx = [1,4,7,9,10] # this is for index
# now we retrieve the data which lies on a particular index
print(arrTwo[idx])
```

```python
# Boolean Masking
temp = arrTwo[arrTwo < 10] # here we are giving a codition 
print(temp)
```
